```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
data=pd.read_excel('myexcel.xlsx')
data
```

```python
team_sizes = data["Team"].value_counts()
team_sizes
```

```python
total_employee=len(data)
```

```python
total_employee
```

```python
team_percetage=(team_sizes/total_employee)*100
```

```python
result=pd.DataFrame({'Team':team_sizes.index,
                     'Number_of_employee':team_sizes.values,
                    'percentage':team_percetage.values},)
result
```

### 2.Segregate the employees w.r.t different positions.

```python
position_dfs = {position: data[data['Position'] == position] for position in data['Position'].unique()}

for position, position_df in position_dfs.items():
    print(f"Employees in {position}:")
    print(position_df)
    print("\n")
```

### 3.Find from which age group most of the employees belong to.

### 4.Find out under which team and position, spending in terms of salary is high.

```python
high_salary=data["Salary"].max()
high_salary
```

```python
high_salary1=data.loc[data['Salary']==high_salary,['Team','Position','Salary']]
high_salary1
```

### 5.Find if there is any correlation between age and salary , represent it visually.

```python
sns.set_style('dark')
plt.scatter(data['Salary'], data['Age'], color='red', marker='o', alpha=0.7)
plt.xlabel('Salary')
plt.ylabel('Age')
plt.title('Correlation Between Age and Salary')
plt.show()
```
