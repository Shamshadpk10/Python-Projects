### Find the outliers

```python
Q1 = data['Salary'].quantile(0.25)
Q3 = data['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data['Salary'][(data['Salary'] < lower_bound) | (data['Salary'] > upper_bound)]

# Visualize the boxplot with outliers
plt.boxplot(data['Salary'])
plt.title(f'Boxplot of Salary')
plt.show()

print(f"Outliers in Salary:")
print(outliers)
```

```python
top_salary=data['Salary'].value_counts().head(15)
```

```python
data['Salary']=data['Salary'].round(2)
data
```

```python
top_salary = data['Salary'].value_counts().head(15).sort_values(ascending=False)
plt.figure(figsize=(15, 6))
sns.barplot(x=top_salary.index, y=top_salary.values)


plt.title('Bar Plot of Top 15 Salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
same_place=data['Place'].value_counts()
same_place
```

### Count the number of people from each place and represent it visually

```python
colors = plt.cm.viridis(np.linspace(0, 1, len(same_place)))
plt.figure(figsize=(15,8))
plt.bar(same_place.index, same_place.values, color=colors)
plt.title("Number Of People From Each Place")
plt.xlabel("Location")
plt.ylabel("Count")

plt.show()
```
